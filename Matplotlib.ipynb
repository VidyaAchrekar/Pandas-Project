### Pie Charts 

```python
labl ='Softwares', 'Scientists','Engineers', 'Managers'
sizes =[100,130,65,25] #total 320
c=['blue', 'green', 'yellow','red']  
plt.pie(sizes,labels=labl,colors=c)
plt.axis('equal')
plt.show()
```

```python
labels ='Softwares', 'Scientists','Engineers', 'Managers'
sizes =[100,130,65,25]
colors=['blue', 'green', 'yellow','red']
d=(0,0.,0.6,1) # distance  
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%' )
plt.axis('equal')
plt.show()
```

```python
labels ='Softwares', 'Scientists','Engineers', 'Managers'
sizes =[100,130,65,25]
colors=['blue', 'green', 'yellow','red']
d=(0,0,0,0.7) # distance  
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%', explode=d)
plt.axis('equal')
plt.show()
```

# Subplots

```python
# Multiple plots in same figure
a = np.arange(1,5)
b = np.arange(6,10)
print (a)
print(b)
plt.figure(1, figsize=(12,10), dpi=400)
plt.subplot(231)
plt.bar(a,b)
plt.subplot(232)
plt.scatter(a, b)
plt.subplot(233)
plt.plot(a, b)
plt.subplot(234)
plt.plot(a, b,'r^')
plt.subplot(235)
plt.plot(a, b, 'gs')
plt.subplot(236)
plt.plot(a, b, 'r-^')

plt.suptitle('Multiple plots')
plt.show()
```
